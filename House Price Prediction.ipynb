```python
def extract_avg_sqft(value):
    if '-' in str(value):
        parts = re.split(r'\s*-\s*', value)
        return (float(parts[0]) + float(parts[1])) / 2
    elif 'Sq. Meter' in str(value):
        parts = re.findall(r'\d+',value)
        return (float(parts[0])*10.764)
    elif 'Perch' in str(value):
        parts = re.findall(r'\d+',value)
        return (float(parts[0])*272.25)
    elif 'Sq. Yards' in str(value):
        parts = re.findall(r'\d+',value)
        return (float(parts[0])*9)
    elif 'Cents' in str(value):
        parts = re.findall(r'\d+',value)
        return (float(parts[0])*435.6)
    elif 'Acres' in str(value):
        parts = re.findall(r'\d+',value)
        return (float(parts[0])*43560)
    elif 'Grounds' in str(value):
        parts = re.findall(r'\d+',value)
        return (float(parts[0])* 2400)
    elif 'Guntha' in str(value):
        parts = re.findall(r'\d+',value)
        return (float(parts[0])* 1089)
    else:
        # numbers = re.findall(r'\d+', str(value))
        # return float(numbers[0]) if numbers else None
        return float(value)
```

```python
df2['avg_total_sqft'] = df2['total_sqft'].apply(extract_avg_sqft)
df2['total_sqft'].unique()
```
